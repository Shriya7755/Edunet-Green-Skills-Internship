## Dataset Overview

- The dataset contains information about renewable energy production and sustainability indicators for different countries.
- It includes solar energy, wind energy, hydropower generation, carbon emissions, renewable energy percentage, and population.
- The dataset will be analyzed to understand relationships between renewable energy production and sustainability metrics.

#2) Check for null values

```python
df.isnull().sum()
```

## Null Value Analysis

- The dataset contains missing values in:
  - Solar_Energy_MW
  - Wind_Energy_MW
  - Renewable_Percentage

- Since these are numerical columns and the missing values are few,
  replacing them with the mean value of their respective columns is a
  reasonable approach.

#3) Handle the null values

```python
df["Solar_Energy_MW"] = df["Solar_Energy_MW"].fillna(df["Solar_Energy_MW"].mean())

df["Wind_Energy_MW"] = df["Wind_Energy_MW"].fillna(df["Wind_Energy_MW"].mean())

df["Renewable_Percentage"] = df["Renewable_Percentage"].fillna(df["Renewable_Percentage"].mean())
```

```python
df.isnull().sum()
```

## Null Value Handling

- Missing values in numerical columns were replaced using the mean value
  of the respective column.
- Mean imputation helps preserve the dataset size and avoids introducing
  extreme values into the analysis.
