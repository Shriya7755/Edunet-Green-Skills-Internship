# Task 2: Python Basics
## 1. List

```python
fruits = ["Apple", "Banana", "Mango"]

fruits.append("Orange")
print("append():", fruits)

fruits.extend(["Kiwi", "Grapes"])
print("extend():", fruits)

fruits.insert(1, "Pineapple")
print("insert():", fruits)

fruits.remove("Banana")
print("remove():", fruits)

fruits.pop()
print("pop():", fruits)
```

##2. Dictionary




```python
print("\n========== DICTIONARY METHODS ==========")

student = {
    "Name": "Shriya",
    "Age": 21,
    "Branch": "Computer Engineering"
}

print("Original Dictionary:", student)

print("keys():", student.keys())

print("values():", student.values())

print("items():", student.items())

print("get('Age'):", student.get("Age"))

student.update({"City": "Ahmedabad"})
print("update():", student)
```

##3. Tuple

```python
print("\n========== TUPLE METHODS ==========")

numbers = (10,20,30,20,40)

print("Tuple:", numbers)

print("count(20):", numbers.count(20))

print("index(30):", numbers.index(30))

print("Length:", len(numbers))

print("Maximum:", max(numbers))

print("Minimum:", min(numbers))
```

##4. Set

```python
print("\n========== SET METHODS ==========")

colors = {"Red","Blue","Green"}

print("Original Set:", colors)

colors.add("Yellow")
print("add():", colors)

colors.update(["Black","White"])
print("update():", colors)

colors.remove("Blue")
print("remove():", colors)

removed = colors.pop()
print("pop():", removed)

print("Current Set:", colors)
```

##5. If Statement

```python
print("\n========== IF ==========")

age = 20

if age >= 18:
    print("Eligible to Vote")
```

##6. If-Else

```python
print("\n========== IF ELSE ==========")

number = 7

if number % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

##7. If-Elif-Else

```python
print("\n----- IF ELIF ELSE -----")

marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

##8. Nested If

```python
print("\n========== NESTED IF ==========")

age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to Vote")
    else:
        print("Not a Citizen")
else:
    print("Minor")
```

##9. Break

```python
print("\n========== BREAK ==========")

for i in range(1,6):

    if i == 4:
        break

    print(i)
```

##10. Continue

```python
print("\n========== CONTINUE ==========")

for i in range(1,6):

    if i == 3:
        continue

    print(i)
```

##11. Pass

```python
print("\n========== PASS ==========")

for i in range(1,4):

    if i == 2:
        pass

    print(i)
```

##14. Len Function

##15. Type Function

```python
print("\n----- TYPE -----")

x = 10
y = 5.5
z = "Python"

print(type(x))
print(type(y))
print(type(z))
```

##16. For Loop

```python
print("----- FOR LOOP -----")

fruits = ["Apple", "Banana", "Mango", "Orange"]

for fruit in fruits:
    print(fruit)
```

##17. While Loop

```python
print("----- WHILE LOOP -----")

count = 5

while count > 0:
    print(count)
    count -= 1

print("Countdown Finished!")
```
